# Threading

```python
import threading
import time
```

```python
def longSquare(num):
    time.sleep(1)
    results[num] = num ** 2

results = {}
t1 = threading.Thread(target=longSquare,args=(1,))
t2 = threading.Thread(target=longSquare,args=(2,))

#Start the Threads
t1.start()
t2.start()

#Wait until threads have completed the execution
t1.join()
t2.join()

print(results)
```

```python
#Using List Comprehensions
def longSquare(num):
    time.sleep(1)
    results[num] = num ** 2

results = {}

#Create Threads
threads = [threading.Thread(target=longSquare,args=(n,)) for n in range(1,11)]

#Start & Join Threads
[t.start() for t in threads]
[t.join() for t in threads]

print(results)
```

# MultiProcessing

```python
# Use Third Party Module Multiprocess
!python3 -m pip install multiprocess
from multiprocess import Process
import time
```

```python
#Using List Comprehensions
def longSquare(num):
    time.sleep(1)
    print(num ** 2)
    print("Finished Computing")

results = {}

#Create Threads
processes = [Process(target=longSquare,args=(n,)) for n in range(1,100)]

#Start & Join Threads
[process.start() for process in processes]
[process.join() for process in processes]

print(results)
```
