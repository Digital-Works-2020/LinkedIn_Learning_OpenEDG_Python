```python
#Magic Method __str__ & __repr__
class Book:
    def __init__(Self, title, author, price):
        super().__init__()
        Self.title = title
        Self.price = price
        Self.author = author

    def __str__(Self):
        return f"{Self.title} by {Self.author} costs ${Self.price}"

    def __repr__(Self):
        return f"Title: {Self.title} - Author: {Self.author} -  Price: ${Self.price}"

b = Book("Think & Grow Rich", "Napoleon Hill", 300)
print(b) #Think & Grow Rich by Napoleon Hill costs $300 
print(str(b)) #Think & Grow Rich by Napoleon Hill costs $300
print(repr(b)) #Title: Think & Grow Rich - Author: Napoleon Hill -  Price: $300
```

```python
#Equality & Comparision Methods
class Book:
    def __init__(Self, title, author, price):
        super().__init__()
        Self.title = title
        Self.price = price
        Self.author = author

    def __eq__(Self, Comp_Obj):
        if not isinstance(Comp_Obj, Book):
            raise ValueError("Cannot compare a book object with other object")
        return Self.title == Comp_Obj.title and Self.author == Comp_Obj.author  and Self.price == Comp_Obj.price

    def __ge__(Self, Comp_Obj):
        if not isinstance(Comp_Obj, Book):
            raise ValueError("Cannot compare a book object with other object")
        return Self.price >= Comp_Obj.price

    def __lt__(Self, Comp_Obj):
        if not isinstance(Comp_Obj, Book):
            raise ValueError("Cannot compare a book object with other object")
        return Self.price <= Comp_Obj.price
        
b1 = Book("Think & Grow Rich", "Napoleon Hill", 39.95)
b2 = Book("War & Peace", "Leo Tolstoy", 29.95)
b3 = Book("Think & Grow Rich", "Napoleon Hill", 39.95)
b4 = Book("To kill a Mocking Bird", "Harper Lee", 24.95)

#Equality Check
print(b1 == b3) #True
print(b1 == b2) #False
try:
    print(b1 == 5)
except ValueError:
    print("Cannot Compare Book with Number")

#Comparision Check
print(b1 >= b2)#True
print(b2 >= b1)#False
print(b1 >= b3)#True
print(b2 <= b1) #True
print(b1 <= b3) #True

#Sorting
books = [b1, b2, b3, b4]
books.sort() #Uses Lesss than by Default. Ascending Order By Default
for book in books:
    print(book.title)
    #Output
    # To kill a Mocking Bird
    # War & Peace
    # Think & Grow Rich
    # Think & Grow Rich

```

```python
#Attribute Access
class Book:
    def __init__(Self, title, author, price):
        super().__init__()
        Self.title = title
        Self.price = price
        Self.author = author
        Self._discount = 0.1
        
    def __str__(Self):
        return f"{Self.title} by {Self.author} costs ${Self.price}"

    def __getattribute__(self, name):
        if name == "price":
            p = super().__getattribute__("price")
            d = super().__getattribute__("_discount")
            return p - (p * d)
        return super().__getattribute__(name)

    def __setattr__(self, name, value):
        if name == "price" and type(value) is not float:
            raise ValueError("Price Must be floating number")
        return super().__setattr__(name, value)

    def __getattribute__(self, name): #Gets called when lookup fails
       return name + " is not here"

b1 = Book("Think & Grow Rich", "Napoleon Hill", float(200))
print(b1.price) #180
print(b1) #Think & Grow Rich by Napoleon Hill costs $180.0
print(b1.non_attr) #non_attr is not here



        
```
